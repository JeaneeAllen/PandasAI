# Data Loading

* Load a CSV file containing at least 500 records and four columns into a regular Pandas data frame and a PandasAI data frame.

```python
np.random.seed(42) # for reproducibility

n = 750 # number of samples

# Generate synthetic data

categories = ["Electronics", "Clothing", "Groceries", "Home", "Beauty"]
regions = ["North", "South", "East", "West"]
payment_methods = ["Credit Card", "Debit Card", "Cash", "Mobile Pay"]

df_synthetic = pd.DataFrame({
    "CustomerID": np.random.randint(1000, 9999, n),
    "TransactionID": np.arange(1, n + 1),
    "Category": np.random.choice(categories, n),
    "Region": np.random.choice(regions, n),
    "PaymentMethod": np.random.choice(payment_methods, n),
    "Amount": np.round(np.random.uniform(5, 500, n), 2)
    })

# Save to CSV
output_path = "synthetic_transactions.csv"
df_synthetic.to_csv(output_path, index=False)
print(f"Synthetic dataset saved to {output_path}")

```

# Load CSV into PandasAI

```python
# Import necessary libraries for PandasAI
import pandasai as pai
from pandasai_litellm import LiteLLM
from openai import OpenAI

client = OpenAI( api_key=api_key) # Initialize the OpenAI client with the API key
```

```python
llm = LiteLLM(
    model="gpt-5-nano", # Specify the model to use
    api_key=api_key
) # Initialize the LiteLLM with the specified model and API key

pai.config.set({
    "llm": llm
}) # Set the LLM for PandasAI
```

```python
df_ai = pai.read_csv("synthetic_transactions.csv") # Load the dataset using PandasAI
print(df_ai.shape) # Check the shape of the dataset loaded with PandasAI
print(df_ai.head()) # Display the first few rows of the dataset loaded with PandasAI
```

➡︎  Two Columns Criteria: Present the code used to filter the data frame based on criteria in two columns using both regular Pandas and PandasAI.
* Regular Pandas: Combine conditions for two columns, such as df[(df['column1'] == 'value1') & (df['column2'] == 'value2')].

* PandasAI: Perform the same operation, adjusting for any differences in handling logical operations.

**Two columns w/ Pandas**

```python
pandas_two = df[
    (df["Category"] == "Electronics") &
    (df["Region"] == "North")
]
pandas_two.head()
```

**Two columns Dataset w/ PandasAI**

```python
ai_two = df_ai.chat(
    "Return a dataframe containing only rows where Category is Electronics and Region is North"
)

ai_two_df = ai_two.value
print(ai_two_df.shape)
print(ai_two_df.head())
```

***
➡︎  Three Columns Criteria: Demonstrate how to filter the data frame using criteria from three columns using both regular Pandas and PandasAI.

 * Regular Pandas: Extend the conditions to three columns, ensuring logical operations are correctly applied, like df[(df['column1'] == 'value1') & (df['column2'] == 'value2') & (df['column3'] == 'value3')].

 * PandasAI: Again, apply these filters, considering any specific nuances of the PandasAI library.

**Three columns Dataset w/ Pandas**

```python
pandas_three = df[
    (df["Category"] == "Electronics") &
    (df["Region"] == "North") &
    (df["PaymentMethod"] == "Credit Card")
]
pandas_three.head()
```

**Three columns Dataset w/ PandasAI**

```python
ai_three = df_ai.chat(
    "Return a dataframe containing only rows where Category is Electronics, Region is North, and Payment_Method is Credit Card"
)

ai_three_df = ai_three.value
print(ai_three_df.shape)
print(ai_three_df.head())
```
